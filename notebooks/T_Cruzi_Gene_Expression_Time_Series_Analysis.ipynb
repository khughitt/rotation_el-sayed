Filtering
---------
In order to reduce the volume of data and remove any genes for which little change in expression was found, we will clip the distribution and keep only the distribution tails, i.e.:  

$$\\{x|x < \mu - 3\sigma\ or \ x > \mu + 3\sigma\\}$$

```python
def col_mask(data, col):
    return ((data[col] < data[col].mean() - (3 * data[col].std())) |
            (data[col] > data[col].mean() + (3 * data[col].std())))
```

```python
bool_mask = np.any([col_mask(data, i) for i in data.columns], axis=0)
```

```python
data = data[bool_mask]
```

```python
data.shape
```

Clustering
----------
Now we are ready to cluster the results. For this initial exploration of the data we can use the [k-means clustering](http://en.wikipedia.org/wiki/K-means_clustering) routine in the [scipy.cluster](http://docs.scipy.org/doc/scipy/reference/cluster.html) package. I tried out several different numbers of clusters to generate and '4' seemed to produce pretty good results.

```python
num_clusters = 4
```

```python
x = data.as_matrix()
```

```python
centroids, _ = kmeans(x, num_clusters)
```

```python
idx, _ = vq(x,centroids)
```

Visualization
-------------
``idx`` now contains an array the same size as our data table. Each cell in the ``idx`` array lists a cluster (from 0 to 3) that the corresponding gene has been assigned to. To add the groupings onto our original data table we first convert it to a pandas [Series](http://pandas.pydata.org/pandas-docs/stable/dsintro.html#series) object and then use the ``DataFrame.join()`` command to append the column to the right of the existing columns.

```python
data = data.join(pandas.Series(idx, index=data.index, name='cluster'))
```

Now we are ready to plot the results of our clustering. For this, we can use the pandas [parallel coordinates](http://pandas.pydata.org/pandas-docs/stable/visualization.html#parallel-coordinates) method.

```python
pandas.tools.plotting.parallel_coordinates(data, 'cluster', colors=("#C8E052", "#47CFEB", "#6B53E0", "#E05382"), use_columns=True)
title("Gene Expression vs. Time")
xlabel("Time (hours)")
ylabel("Relative Expression")
```

RNA-seq contigs mapped to T. cruzi
==================================
Next, we will take an initial look at T. cruzi expression data. Since there is currently no control sequence data available, we will only compare replicates at different time points.

```python
# Read in data
col_names = ['gene', 4, 6, 12, 20, 24, 48, 72]

replicates = pandas.read_csv('../../data/tcruzi_counts_replicates', 
                             index_col=0, skiprows=1, header=None, names=col_names)

# Currently we don't have any control DEseq counts for T. cruzi
data = replicates / replicates.sum().astype('float')

# filter out rows which don't show any significant change in transcription
def col_mask(data, col):
    return ((data[col] < data[col].mean() - (3 * data[col].std())) |
            (data[col] > data[col].mean() + (3 * data[col].std())))

# create a boolean mask with all rows that have no significant values
# set to False
bool_mask = np.any([col_mask(data, i) for i in data.columns], axis=0)
data = data[bool_mask]

# k-means clustering
# http://glowingpython.blogspot.com/2012/04/k-means-clustering-with-scipy.html
num_clusters = 4
data_matrix = data.as_matrix()
#data_matrix = whiten(data_matrix)
centroids, distortion = kmeans(data_matrix, num_clusters)
idx, distortion = vq(data_matrix, centroids)

# add column with cluster grouping at end of data
data = data.join(pandas.Series(idx, index=data.index, name='cluster'))

# create parallel coordinates plot
plt.figure()
parallel_coordinates(data, 'cluster', use_columns=True,
                     colors=('#556270', '#4ECDC4', '#C7F464', '#FF6B6B'))
title("Gene Expression vs. Time")
xlabel("Time (hours)")
ylabel("Relative Expression")
```
